TOP 4 FINISHES

```python
top_4_countries =pd.concat([df['Winner'],df['Runners-Up'],df['Third'],df['Fourth']])
top_4_countries= top_4_countries.value_counts()
print("Number of times countries finished in top 4:")
print(top_4_countries)
```

```python
PLOTTING TOP 4 FINISHES
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x = top_4_countries.index,y = top_4_countries.values,palette='muted')
plt.title('Top 4 Finishes By Country(1930-2014)')
plt.xlabel('Country')
plt.ylabel('Top 4 Finishes')
plt.xticks(rotation=45)
plt.show()
```

PLOTTING THE ATTENDANCE OF EACH WORLD CUP

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df,x='Year',y='Attendance',marker='o',color='r')
plt.title('Attendance of Each World Cup(1930-2014)')
plt.xlabel('Year')
plt.ylabel('Attendance')
plt.grid(True)
plt.show()
```
